```python
import json
```

```python
#Function to read file data.
def loadData(file):
    with open (file, "r") as f:
        data = json.load(f)
    return data
```

```python
d = loadData("data.json")
d
```

```python
#Function to display user relations and liked pages
def displayUserData(data):
    print("Users and their connections::\n")
    for user in data['users']:
        print(f"{user['id']} {user['name']} is friends with:{user['friends']} and likes these pages: {user['liked_pages']}")
    print("\n")
    print("Page Info:\n")
    for page in data['pages']:
        print(f"Page ID: {page['id']} and Page Name: {page['name']}")

displayUserData(d)
```
